```python
import matplotlib.pyplot as plt
import pyemma
import numpy as np
```

```python
f1=open('sT_sR_w_time_replica_all.dat','r')
pc1 = np.zeros(2403)   # pc1
pc2 = np.zeros(2403)   # pc2
for i in range(2403):
    s = f1.readline()
    token = s.split()
    pc1[i]=token[1]
    pc2[i]=token[2]
```

```python
import matplotlib.pyplot as plt
import pyemma

plt.rcParams["font.weight"] = "bold"

# Create a new figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Assuming pc1 and pc2 are already defined
pyemma.plots.plot_free_energy(pc1, pc2, ax=ax, legacy=False, cmap='jet', nbins=50)

# Set the linewidths of the spines
ax.spines["top"].set_linewidth(5)
ax.spines["left"].set_linewidth(5)
ax.spines["right"].set_linewidth(5)
ax.spines["bottom"].set_linewidth(5)

# Adjust the position of the plot within the figure
plt.gcf().subplots_adjust(bottom=0.23)
plt.gcf().subplots_adjust(left=0.3)

# Set the labels for the axes
ax.set_xlabel('$\mathbf{S_T}$' , size=30, fontweight='bold')
ax.set_ylabel('$\mathbf{S_R}$', size=30, fontweight='bold')

# Set tick parameters
#ax.tick_params(axis="both", length=15, width=4, color="black")
ax.tick_params(axis="x", labelsize=15, labelrotation=0)
ax.tick_params(axis="y", labelsize=15, labelrotation=0)
ax.tick_params(axis="x", which="minor", bottom="off")



# Show the plot
plt.show()

# Save the plot
fig.savefig('fes_1kx5_cg.png', dpi=600)
#fig.savefig('breathing_distance.eps', dpi=600)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
#import matplotlib.ticker
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
import math
from scipy.optimize import curve_fit
from scipy import stats
from pylab import *
import warnings
#warnings.filterwarnings('ignore')
from sklearn.metrics import r2_score
```

```python
n = 1202
time = np.zeros(n)
corr = np.zeros(n)
time1 = np.zeros(n)
corr1 = np.zeros(n)
f1 = open('1KX5_backmap_histone_fit_transloc_rot_resd.out','r')
#f2 = open('fit.dat','w')
z1 = []
a1 = []
a2 = []
z2 = []
```

```python
for i in range(n):
    s = f1.readline()
    token = s.split()
    time[i] = token[0]
    corr[i] = token[1]
    if (time[i] > 0.2 and time[i] < 2.0):
        a1.append(float(time[i])) 	
        z1.append(float(corr[i]))
    if (time[i] > 2.6 and time[i] < 4.6):
        a2.append(float(time[i])) 	
        z2.append(float(corr[i])) 
```

```python
def fitstraight(t,A,B,C):
	return A*t**B+C
parameters1, covariance1 = curve_fit(fitstraight, a1, z1)
fit_1 = parameters1[0]
fit_2 = parameters1[1]
fit_3 = parameters1[2]
corr_pred = fitstraight(a1, *parameters1)
a = r2_score(z1, corr_pred)
print('R2=',a)
print('A=',fit_1)
print('B=',fit_2)
print('C=',fit_3)
parameters2, covariance2 = curve_fit(fitstraight, a2, z2)
fit_4 = parameters2[0]
fit_5 = parameters2[1]
fit_6 = parameters2[2]
corr_pred1 = fitstraight(a2, *parameters2)
a1 = r2_score(z2, corr_pred1)
print('R2=',a1)
print('A=',fit_4)
print('B=',fit_5)
print('C=',fit_6)
plt.rcParams["font.weight"] = "bold"
#plt.rcParams["axes.labelweight"] = "bold"
fig, ax = plt.subplots(figsize=(9,7), dpi=100)
#fig, ax = plt.subplots(111)
ax.spines["top"].set_linewidth(8)
ax.spines["left"].set_linewidth(8)
ax.spines["right"].set_linewidth(8)
ax.spines["bottom"].set_linewidth(8)

plt.gcf().subplots_adjust(bottom=0.23)
plt.gcf().subplots_adjust(left=0.3)
ax.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))

ax.scatter(time, corr,color="black",marker = '^',facecolors="none",edgecolors='black',linewidth=1, s=30) # for inset
ax.plot(time, fitstraight(time, *parameters1), color="green",linewidth=5,linestyle='dashed',label="Fitted line")
ax.plot(time, fitstraight(time, *parameters2), color="green",linewidth=5,linestyle='dashed',label="Fitted line")

ax.tick_params(axis="both", direction="in", length=15, width=3, color="black")
ax.tick_params(axis="x", labelsize=30)
ax.tick_params(axis="y", labelsize=30)
#ax.tick_params(axis="x", labelsize=50, labelrotation=0)
#ax.tick_params(axis="y", labelsize=50, labelrotation=0)
ax.tick_params(axis = 'x',which='minor', direction='in',length=10, width=4,color='black')
ax.tick_params(axis = 'y',which='minor', direction ='in', length=10, width=4,color='black')
# For inset no label
ax.set_ylabel(r"$S_T$" , size = 30 ,fontweight='bold')   # For inset no label
#ax.set_ylabel(r"$J^{H^N-C_\alpha}_{N-H_\alpha;T7/K6}(\Delta t)$", size = 35,fontweight='bold') # CHange for residue
ax.set_xlabel(r"$\Delta t (microsecond)$", size = 30,fontweight='bold')
ax.get_legend()
ax.legend(loc="upper right",fontsize=10,frameon=False)
plt.show()
fig.savefig('fit-bi.eps',dpi=600) # change name
fig.savefig('fit-bi.png',dpi=600)
```
