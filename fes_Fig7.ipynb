```python
from sklearn.preprocessing import StandardScaler
import numpy as np
# Standardize the data
scaler = StandardScaler()
standardized_data = scaler.fit_transform(combined_df)
from sklearn.decomposition import PCA
# Apply PCA
pca = PCA(n_components=2)  # Adjust the number of components as needed
principal_components = pca.fit_transform(standardized_data)
import matplotlib.pyplot as plt
# Explained variance
explained_variance = pca.explained_variance_ratio_
# Plotting the principal components
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Distance and Angle Data')
plt.show()
# PC1 and PC2
pc1 = principal_components[:, 0]
pc2 = principal_components[:, 1]
plt.rcParams["font.weight"] = "bold"
# Define the number of bins
num_bins = 50
# Create 2D histogram
hist, xedges, yedges = np.histogram2d(pc1, pc2, bins=num_bins, density=True)
# Centers of the bins
xcenters = 0.5 * (xedges[1:] + xedges[:-1])
ycenters = 0.5 * (yedges[1:] + yedges[:-1])
X, Y = np.meshgrid(xcenters, ycenters)
# Ensure no zero values in histogram to avoid log(0)
hist = np.where(hist == 0, np.nan, hist)
# Calculate free energy
free_energy = -np.log(hist)
# Replace NaNs with a large number to indicate very high free energy (e.g., areas with no data)
free_energy = np.nan_to_num(free_energy, nan=np.inf)
# Plotting the free energy
plt.figure(figsize=(12, 8))
plt.contourf(X, Y, free_energy, levels=100, cmap='jet')
plt.colorbar(label='Free Energy/kT')
plt.xlabel('PC1', size=30, fontweight='bold')
plt.ylabel('PC2', size=30, fontweight='bold')
#plt.title('Free Energy Landscape')
ax = plt.gca()
ax.spines['top'].set_linewidth(5)
ax.spines['right'].set_linewidth(5)
ax.spines['bottom'].set_linewidth(5)
ax.spines['left'].set_linewidth(5)
ax.tick_params(axis="x", labelsize=15, labelrotation=0)
ax.tick_params(axis="y", labelsize=15, labelrotation=0)
ax.tick_params(axis="x", which="minor", bottom="off")
plt.savefig('pca_lz0_CG_pcacode.png', dpi=600)
plt.show()
```
