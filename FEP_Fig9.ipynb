```python
import numpy as np
import matplotlib.pyplot as plt

# Load the free energy landscape data
# Assuming the columns are: x, y, z (free energy)
free_energy_data = np.loadtxt('fes_CG_lz0.dat')
x = free_energy_data[:, 0]
y = free_energy_data[:, 1]
z = free_energy_data[:, 2]

# Load the minimum energy path data
# Assuming the columns are: x, y, z (along the path)
path_data = np.loadtxt('cg_lz0_min_path.dat')
x_path = path_data[:, 0]
y_path = path_data[:, 1]
z_path = path_data[:, 2]  # We might not need z_path for a 2D contour plot

# Create the contour plot
fig, ax = plt.subplots(figsize=(10, 8))

# Plot the free energy surface as a contour plot
contour = ax.tricontourf(x, y, z, levels=10, cmap='rainbow')

# Overlay the minimum energy path
ax.plot(x_path, y_path, color='red', linewidth=2.5, label='Minimum Energy Path')

# Set labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Add a colorbar for the contour plot
cbar = plt.colorbar(contour)
cbar.set_label('Free Energy')

# Add a legend
ax.legend()

# Show the plot
plt.show()

# Optionally, save the plot
fig.savefig('free_energy_contour_with_minimum_path.png', dpi=300)
```

```python
f4=open('ST_SR_Cg_all_lz0.dat','r')
pc1 = np.zeros(4191)   # pc1
pc2 = np.zeros(4191)   # pc2
for i in range(4191):
    s = f4.readline()
    token = s.split()
    pc1[i]=token[1]
    pc2[i]=token[2]
import matplotlib.pyplot as plt
import pyemma

plt.rcParams["font.weight"] = "bold"

# Create a new figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Assuming pc1 and pc2 are already defined
pyemma.plots.plot_free_energy(pc1, pc2, ax=ax, legacy=False, cmap='jet', nbins=40)

# Overlay the minimum energy path
ax.plot(x_path, y_path, color='black', linewidth=3.0, linestyle='--', label='Minimum Energy Path')

# Set the linewidths of the spines
ax.spines["top"].set_linewidth(5)
ax.spines["left"].set_linewidth(5)
ax.spines["right"].set_linewidth(5)
ax.spines["bottom"].set_linewidth(5)

# Adjust the position of the plot within the figure
plt.gcf().subplots_adjust(bottom=0.23)
plt.gcf().subplots_adjust(left=0.3)

# Set the labels for the axes
ax.set_xlabel('$\mathbf{S_T}$' , size=30, fontweight='bold')
ax.set_ylabel('$\mathbf{S_R}$', size=30, fontweight='bold')

# Set tick parameters
#ax.tick_params(axis="both", length=15, width=4, color="black")
ax.tick_params(axis="x", labelsize=15, labelrotation=0)
ax.tick_params(axis="y", labelsize=15, labelrotation=0)
ax.tick_params(axis="x", which="minor", bottom="off")



# Show the plot
plt.show()

# Save the plot
fig.savefig('fes_lz0_cg.png', dpi=600)
#fig.savefig('breathing_distance.eps', dpi=600)
```
